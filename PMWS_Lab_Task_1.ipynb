```python
!pip install numpy matplotlib
import numpy as np
import matplotlib.pyplot as plt

def generate_am_signal(Vc, fc, Vm, fs, t):
    m = Vm / Vc  # Modulation Index
    carrier = Vc * np.sin(2 * np.pi * fc * t)
    modulating_signal = Vm * np.sin(2 * np.pi * fs * t)
    am_signal = Vc * (1 + m * np.sin(2 * np.pi * fs * t)) * np.sin(2 * np.pi * fc * t)
    return am_signal, carrier, modulating_signal

# Parameters
Vc = float(input("Enter carrier voltage Vc: "))
fc = float(input("Enter carrier frequency fc (Hz): "))
Vm = float(input("Enter message voltage Vm: "))
fs = float(input("Enter message frequency fs (Hz): "))
duration = float(input("Enter signal duration (s): "))

# Time array
t = np.linspace(0, duration, int(1000 * duration))

# Generate signals
am_signal, carrier, modulating_signal = generate_am_signal(Vc, fc, Vm, fs, t)

# Plot the signals
plt.figure(figsize=(10, 8))

plt.subplot(3, 1, 1)
plt.plot(t, carrier, label='Carrier Signal')
plt.title('Carrier Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.legend()

plt.subplot(3, 1, 2)
plt.plot(t, modulating_signal, label='Message Signal')
plt.title('Message Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.legend()

plt.subplot(3, 1, 3)
plt.plot(t, am_signal, label='AM Signal')
plt.title('Amplitude Modulated Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.legend()

plt.tight_layout()
plt.show()
```
